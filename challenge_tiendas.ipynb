### Importación de datos



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
####----------------------------------------------------------------------------------------------------------####
### IMPORTANTE PARA REALIZACION DEL CHALLENGE VAMOS A SUPONER QUE LOS VALORES QUE REFLEJAN DINERO SERAN EN COP ###
####----------------------------------------------------------------------------------------------------------####

### en lo siguiente vamos a agregar a cada "tienda" una nueva columna que se llama 'tienda' esto con el fin de poder analizar datos en conjunto de
# cada tienda ###
tienda['tienda'] = 'tienda 1'
tienda2['tienda'] = 'tienda 2'
tienda3['tienda'] = 'tienda 3'
tienda4['tienda'] = 'tienda 4'

## y ahora con ps.concat combinamos los Dataframes de todas las tiendas y lo llamamos df_tiendas ##
df_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

##--------------------------------------------------------------------------------------------------------------------------------##
### VOY A DEJAR UN EJEMPLO COMENTADO DE COMO SE VERIA EL TOTAL DE FACTURACION SI NO COMBINARA TODOS LOS DATAFRAMES EN df_tiendas ###
##--------------------------------------------------------------------------------------------------------------------------------##


# ### primero necesitamos acceder a los valores del dataframe ###t
# columna_precio = tienda['Precio'] ## primero accedemos a la columna del dataframe para poder acceder a los valores por fila ##


# total_facturacion = columna_precio.sum() # como fila es una tupla indice valor no al pueda sumar, entonces para poder realizar la suma
# # simplemente accedemos o usamos la columna_precio#

# print(f"el total de la facturacion de la tienda 1 es {total_facturacion}")

# ## necesitamos sumar los valores que se encutran registramos en la columna Precios para poder saber total de facturacion ##

# ##----------------------------------------------------------------------------------------------------------------------------------------------##

# columna_precio2 = tienda2['Precio']

# total_facturacion2 = columna_precio2.sum()

# print(f"el total de facturacion de la tienda 2 es: {total_facturacion2}")

# ##----------------------------------------------------------------------------------------------------------------------------------------------##

# columna_precio3 = tienda3['Precio']

# total_facturacion3 = columna_precio3.sum()

# print(f"el total de facturacion de la tienda 3 es: {total_facturacion3}")

# ##----------------------------------------------------------------------------------------------------------------------------------------------##

# columna_precio4 = tienda4['Precio']

# total_facturacion4 = columna_precio4.sum()

# print(f"el total de facturacion de la tienda 4 es: {total_facturacion4}")

# total_facturacion_tiendas = (total_facturacion+total_facturacion2+total_facturacion3+total_facturacion4)

# print(f"el total de facturacion de todas las tiendas es: {total_facturacion_tiendas}")
# ##----------------------------------------------------------------------------------------------------------------------------------------------##

## y esto seria usando df_tiendas ##
total_facturacion_por_tienda = df_tiendas.groupby('tienda')['Precio'].sum().reset_index(name='Total Facturación')

total_facturacion_por_tienda.head()
```

# 2. Ventas por categoría

```python
from os import name


## despues de tener todos los dataframes en una sola variable agrupamos con el metodo .groupby el df_tiendas con las categorias de los productos qu
# que el dataframe se llama Categoria del Producto despues con el metodo .size contamos cuantas veces aparece cada categoria en cada tienda y ya al final
# con el metodo .reset_index(name='Ventas') convierte los índices a columnas normales y nombra la nueva columna ventas##
ventas_por_categoria = df_tiendas.groupby('Categoría del Producto').size().reset_index(name='Ventas')


ventas_por_categoria.head()
```

# 3. Calificación promedio de la tienda


```python


##----------------------------------------------------------------------------------------------------------------------------##
# lo anterior se hizo para generar una columna del nombre de la tienda para poder usar el groupby de ejemplo tienda1 con la Calificación y arroje un promedio#
##----------------------------------------------------------------------------------------------------------------------------##

# lo siguiente se hizo tomando de df_tiendas que es un Dataframe de todas las tiendas accedemos a la columna Calificación y decimos que en la siguiente columna
#lo que no sea valores numeros con pd.to_numeric los covienta en NaN#
df_tiendas['Calificación'] = pd.to_numeric(df_tiendas['Calificación'], errors='coerce')
#------------------------------------------------------------------------------------------------------------------------------#

# como ya tenemos una columna en cada Df de las tiendas ppodemos agrupalas y realizar el resultado del promedio de calificacion por tienda #
promedio_calificacion = df_tiendas.groupby('tienda')['Calificación'].mean().round(2).reset_index(name='Calificación promedio')

# print(f"la calificacion promedio de la tienda es: {promedio_calificacion}")
promedio_calificacion.head()

# df_tiendas['tienda'].value_counts()
```

# 4. Productos más y menos vendidos

```python
# # En este paso, debes identificar qué productos fueron los más vendidos y los menos
#  vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
##-----------------------------------------------------------------------------------------------------------##
ventas_producto = df_tiendas.groupby(['tienda','Producto']).size().reset_index(name='ventas')
#primero se creo una varible para para agrupar tienda con producto que con .size() va a contar las columnas que agrupas que en este caso fueorn dos (tienda y producto)
#y va a contar las filas por esa columnna#


## mas_vendido_por_tienda.loc (el punto loc usa los indices) toma ventas_producto y lo usa como indice para luego con el groupby danos por cada tienda el producto mas vendido
#con el idxmax ##
mas_vendido_por_tienda = ventas_producto.loc[
    ventas_producto.groupby('tienda')['ventas'].idxmax()
]

print("mas vendido por tiendas \n")
mas_vendido_por_tienda.head()
```

```python
## en el siguiente apartado se usa mismos indices pero on el idxmin nos dal el menor producto vendido ##
menos_vendido_por_tienda = ventas_producto.loc[
    ventas_producto.groupby('tienda')['ventas'].idxmin()
]
print("menos vendido por tiendas \n")
menos_vendido_por_tienda.head()
```

# 5. Envío promedio por tienda

```python
##En este paso, debes calcular el costo de envío promedio para cada tienda.
#El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda ##

costo_envio_por_tienda = df_tiendas.groupby('tienda')['Costo de envío'].mean().round(1).reset_index(name='Costo de envío promedio')

print(f"el costo de envio promedio de las tiendas es:\n")
costo_envio_por_tienda.head()
```

# 6. Rentabilidad de las tiendas



```python
from sklearn.preprocessing import MinMaxScaler
#importamos metodos muy importantes para poner normalizar los valores

df_rentabilidad = pd.merge(total_facturacion_por_tienda, costo_envio_por_tienda, on='tienda')
##con el indice de tienda combinamos total_facturacion_... y costo_envio_... para que esto nos arroje un df de 3 columnas##

scaler = MinMaxScaler()
##creamos una instacion del metodo MinMaxScaler para usarla de manera mas facil ##

##creamos dos nuevas columnas para el nuevo df que tiene facturacion y costo de envios y despues el scaler.fit_transform aplica la normalizacion de los valores
# que se encuentras en las primeras columnas y los guarda en las nuevas columnas ya normalizados##
df_rentabilidad[['facturacion_normalizada','Envios normalizados']] = scaler.fit_transform(df_rentabilidad[['Total Facturación', 'Costo de envío promedio']])
##la normalizacion se hace con el fin de que si por ejemplo en este caso que facturacion tiene valores mas grandes pesen mas que los valores de costos de envios##


df_rentabilidad['Rentabilidad'] = ((0.7 * df_rentabilidad['facturacion_normalizada']) +  0.3 * ( 1 -  df_rentabilidad['Envios normalizados']))
#creamos al final una columna de Rentabilidad donde le damos un peso del 70% a la facturacion y un 30% a los envios que sean con mas bajo costo se pone el
# 1 - para que la normalizacion funcione de manera inversa#

## al final usamos round(2) para que los valores normalizados sean mas cortos, se redondeen y sea mas claro al momento de visualizar el Dataframe ##
df_rentabilidad = df_rentabilidad.round(2)

df_rentabilidad.head()
```

# 7. GRAFICOS


```python
import matplotlib.pyplot as plt
import seaborn as sns

## Grafica para mostrar el porcentaje que aporta cada tienda a la facturacion total ##

#datos
labels = total_facturacion_por_tienda['tienda']
values = total_facturacion_por_tienda['Total Facturación']

# Definimos el tamaño del grafico
plt.figure(figsize=(6, 6)) # (Ancho, Alto) en pulgadas
# Creamos el gráfico de pastel (pie chart)
plt.pie(
    values,                 # Los valores que determinan el tamaño de cada porción del pastel
    labels=labels,          # Las etiquetas que se mostrarán en cada porción
    autopct='%1.1f%%',      # Formato para mostrar el porcentaje con 1 decimal
    startangle=140          # Ángulo inicial para girar el gráfico y hacerlo más legible
)
plt.title('Participación de cada tienda en la facturación total')
plt.axis('equal')  # Para que sea un circulo perfecto
plt.show()
```

```python

## vamos a crear grafico de ventas por productos ##

# creamos los datos de labels y values
labels = ventas_por_categoria['Categoría del Producto']
values = ventas_por_categoria['Ventas']

colors = sns.color_palette("pastel")

plt.figure(figsize=(6,6)) #damos tamaño a la grafica
plt.bar(labels, values, color=colors) #decimos cuales van a ser los valores que va a usar la grafica
plt.xticks(rotation=45) #rotamos un poco los nombre de los productos
plt.title('Ventas por categoría') #damos titulo de la grafica
plt.xlabel('Categoría del Producto') #damos indices y valores en el eje X
plt.ylabel('Ventas') # damos indices y valores en el eje Y
plt.show() # y con .show() mostramos la grafica renderiza datos y crea graficos
```
